```python
import pandas as pd
```

```python
PATH_FILE_SCORE_FILTRO_ALGORITMO = './scoreFiltratiFinaliCsvSuperFiltraggio.csv'
PATH_FILE_SCORE_FILTRO_MANUALE = './scoreFiltratiFinaliExcelSuo.csv'
PATH_FILE_SCORE_FULL = './scoreCsvDatasetFinale.csv'

scores_filtro_algoritmo = pd.read_csv(PATH_FILE_SCORE_FILTRO_ALGORITMO)
scores_filtro_manuale = pd.read_csv(PATH_FILE_SCORE_FILTRO_MANUALE)
scores_filtro_full = pd.read_csv(PATH_FILE_SCORE_FULL)
```

```python
def calcola_id_soggetto_da_utente(u):
    return u.split("_")[0]

def calcola_num_acquisizione_da_utente(u):
    return int(u.split("_")[1])
```

```python
scores_filtro_full['Primo soggetto'] = scores_filtro_full['Utente1'].apply(lambda u: calcola_id_soggetto_da_utente(u))
scores_filtro_full['Secondo soggetto'] = scores_filtro_full['Utente2'].apply(lambda u: calcola_id_soggetto_da_utente(u))
```

```python
set_filtro_algoritmo = set(list(scores_filtro_algoritmo.Utente1.unique()) +  list(scores_filtro_algoritmo.Utente2.unique()))
set_filtro_manuale = set(list(scores_filtro_manuale.Utente1.unique()) + list(scores_filtro_manuale.Utente2.unique()))
set_full = set(list(scores_filtro_full.Utente1.unique()) + list(scores_filtro_full.Utente2.unique()))
```

```python
filtrati_filtro_algoritmo = set(set_full - set_filtro_algoritmo)
filtrati_filtro_manuale = set(set_full - set_filtro_manuale)
```

```python
df_acquisizioni = pd.DataFrame()
for acquisizione in filtrati_filtro_manuale:
    df_acquisizione = scores_filtro_full[(scores_filtro_full.Utente1 == acquisizione) |  (scores_filtro_full.Utente2 == acquisizione)]
    df_acquisizione = df_acquisizione[df_acquisizione['Primo soggetto'] == df_acquisizione['Secondo soggetto']] 
    media_score = df_acquisizione.ScoreML.mean()
    min_score = df_acquisizione.ScoreML.min()
    perc_min_score = len(df_acquisizione[df_acquisizione.ScoreML <= 0.25]) / len(df_acquisizione)
    riga_df = pd.DataFrame(data={'acquisizione': [acquisizione], 'score_medio': [media_score], 'perc_min_score': [perc_min_score]})
    df_acquisizioni = pd.concat([df_acquisizioni, riga_df])
```

```python
a = df_acquisizioni.sort_values(by='score_medio')
a = a.reset_index(drop=True)
```
