```python
import pandas as pd
```

```python
PATH_CSV_LABELS = 'tabellaPseudoLabelsDatasetCompleto.csv'
PATH_EXCEL_SUO_LABELS = '../Nuovi Dati/Dataset_Sensori_Da_Mat.csv'
```

```python
df_labels = pd.read_csv(PATH_CSV_LABELS)
df_labels
```

```python
df_excel_suo = pd.read_csv(PATH_EXCEL_SUO_LABELS)
lista_acquisizioni_filtrate = set(df_excel_suo.Acquisizione1.unique()) | set(df_excel_suo.Acquisizione2.unique())
lista_acquisizioni_filtrate = [i.replace('.dat', '') for i in lista_acquisizioni_filtrate]
```

```python
lista_acquisizioni = list(set(df_labels.Acquisizione.unique()) - set(lista_acquisizioni_filtrate))

PATH_FILE_SCORE = '../scoreCsvDatasetFinale.csv'
scores = pd.read_csv(PATH_FILE_SCORE)
scores = scores[~scores.Utente1.isin(lista_acquisizioni)]
scores = scores[~scores.Utente2.isin(lista_acquisizioni)]
scores
```

```python
scores.to_csv('../scoreFiltratiFinaliExcelSuo.csv' ,index=False)
```
